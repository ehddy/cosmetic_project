## 2. 클린 뷰티 기준



1. 조건 
    1. 올리브영 선정 위험도 성분 개수 (올리브영 기준) (단위: 개)

    3. 동물 실험을 하지 않는 브랜드 여부(단위: 1, 0)

    5. 천연 유기농 성분 개수(단위: 개)  
    6. 공병 수거여부 (단위: 1, 0)
      

2. 등급 부여 기준(기본 : 시나리오 1)
    1. 모든 조건  충족: 1등급 = 클린 뷰티 

    2. 1~2개 조건 제외 모두 충족: 2등급  

    3. 3~4개 조건 제외 모두 충족: 3등급  

    4. 5~6개 조건 제외 모두 충족: 4등급
    


## 3. 클린 뷰티 등급 부여 함수 구현

```python
# 조건 함수
def A(x):
    if x == 0:
        return 1
    else:
        return 0 
def B(x):
    if x == 0:
        return 1
    else:
        return 0 
def C(x):
    if x == 1:
        return 1
    else:
        return 0
def D(x):
    if x == 0:
        return 1
    else:
        return 0
def E(x):
    if x >= 1:
        return 1
    else:
        return 0
def F(x):
    if x == 1:
        return 1
    else:
        return 0
```

```python
# 충족 여부 
def clean_beauty_y_or_n(data):

    data['A 조건 충족 여부'] = (data['올리브영 선정 유해 성분 개수'].apply(lambda x: A(x))) 
    data['B 조건 충족 여부'] = (data['높은 위험도 성분 개수'].apply(lambda x: B(x)))
    data['C 조건 충족 여부'] = (data['동물실험을 하지 않는 브랜드'].apply(lambda x: C(x))) 
    data['D 조건 충족 여부'] = (data['동물성 성분 개수'].apply(lambda x: D(x)))
    data['E 조건 충족 여부'] = (data['천연 및 유기농 성분 개수'].apply(lambda x: E(x))) 
    data['F 조건 충족 여부'] = (data['공병 수거를 하는 브랜드'].apply(lambda x: F(x)))
    
    return data
```

```python
# 시나리오 1: 모든 조건이 충족되었을 경우 
def rating_1(x):
    if x == 6:
        return '1등급'
    elif x >= 4:
        return '2등급'
    elif x >= 2:
        return '3등급'
    else:
        return '4등급'   
def clean_beauty_grade_SNCN1(data):

    data['조건 충족 개수'] = (data['A 조건 충족 여부']) + (data['B 조건 충족 여부'])  + (data['C 조건 충족 여부']) + (data['D 조건 충족 여부']) + (data['E 조건 충족 여부']) + (data['F 조건 충족 여부'])
    
    data['클린 뷰티 등급'] = data['조건 충족 개수'].apply(lambda x : rating_1(x))    
    return data
```

```python
# 시나리오 2: 브랜드 여부에 가산점을 주는 방식으로 

def rating_2(x):
    if x >= 6:
        return '1등급'
    elif x >= 4:
        return '2등급'
    elif x >= 2:
        return '3등급'
    else:
        return '4등급'


def clean_beauty_grade_SNCN2(data):

    data['조건 충족 개수'] = (data['A 조건 충족 여부']) + (data['B 조건 충족 여부']) + (data['C 조건 충족 여부']*2) + (data['D 조건 충족 여부']) + (data['E 조건 충족 여부']) + (data['F 조건 충족 여부'] *2)
    
    (data['C 조건 충족 여부']) + (data['F 조건 충족 여부'])
    
    data['클린 뷰티 등급'] = data['조건 충족 개수'].apply(lambda x : rating_2(x))    
    
    return data
```

```python
# 시나리오 3: 비건을 위한 화장품 유해성분(올리브영, EWG) x, 동물성 성분 x일때, -> 등급이 아닌 인증 마크 
def rating_3(x):
    if x == 3:
        return 1
    else:
        return 0


def clean_beauty_grade_SNCN3(data):

    data['조건 충족 개수'] = (data['A 조건 충족 여부']) + (data['B 조건 충족 여부']) + (data['D 조건 충족 여부']) 
    
    data['비건 인증'] = data['조건 충족 개수'].apply(lambda x : rating_3(x))    
    
    return data
```

```python
# 시나리오 4: 비건을 위한 화장품 유해성분(올리브영, EWG) x, 동물 실험 여부 x  -> 등급이 아닌 인증 마크 
def rating_4(x):
    if x == 3:
        return 1
    else:
        return 0


def clean_beauty_grade_SNCN4(data):

    data['조건 충족 개수'] = (data['A 조건 충족 여부']) + (data['B 조건 충족 여부']) + (data['C 조건 충족 여부']) 
    
    data['크루얼티프리 인증'] = data['조건 충족 개수'].apply(lambda x : rating_4(x))    
    
    return data
```

```python
data = clean_beauty_y_or_n(data)
data = clean_beauty_grade_SNCN4(data)
```
