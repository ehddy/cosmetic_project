```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd


from selenium.webdriver.common.by import By
from datetime import datetime
```

```python
def CoosCrawling():
    
    page_num = int(input('몇 페이지까지 추출하시겠습니까? '))
    
    driver = webdriver.Chrome('/Users/dongyokim/main/jupyter/chromedriver/chromedriver')
    driver.set_window_size(1000, 1000)
    
    total_ingredient_list = []
    
    
    print('#####Coos 크롤링을 시작합니다. ######')
    print()
    
    
    for i in range(1, page_num + 1):
        driver.get('https://coos.kr/products?page={}'.format(i))
        
        time.sleep(0.4)
    
        row_list = driver.find_elements(By.CSS_SELECTOR, '#__next > div > div > main > div > div > div > div.twelve.wide.column > div > div:nth-child(3) > div:nth-child(1) > div > div > div.MuiTableContainer-root > table > tbody > tr')
        
        time.sleep(0.4)
        
        for row_num in range(0, len(row_list), 2):
            row = row_list[row_num] 

            # 제품명 
            ingredient_name = row.find_element(By.CSS_SELECTOR, 'th > strong').text

            # 제품 카테고리 
            ingredient_categoty = row.find_element(By.CSS_SELECTOR, 'td:nth-child(4)').text

            # 제품 기능 
            ingredient_function = row.find_element(By.CSS_SELECTOR, 'td:nth-child(5)').text

            total_ingredient_list.append([ingredient_name, ingredient_category, ingredient_function])

    data = pd.DataFrame(data=total_ingredient_list, columns=['제품명', '종류', '제품 기능'])    
    print('####### Coos 크롤링을 종료하겠습니다. ########')
    
    driver.quit()
    
    return data
```
